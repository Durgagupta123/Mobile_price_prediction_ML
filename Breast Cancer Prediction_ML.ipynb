```python
#Now select the input and output
x = df.drop("diagnosis",axis=1)    #input selected
y = df["diagnosis"]    #output selected
```

```python
#Split the model into 70-30% 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#To check the number of rows and columns in train - test split
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
#Scale the model by inbuilt class StandardScaler from sklean package
from sklearn.preprocessing import StandardScaler

#create an object for StandardScaler class
ss = StandardScaler()
```

```python
#apply on training and testing data
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
#To check 
x_train 
```

```python
x_test
```

```python
#Now we apply Classification Algorithm to our dataset
#we create the user define function
def create_model(model):
    model.fit(x_train,y_train)   #to train the dataset into 70% data
    y_pred = model.predict(x_test)  #to test the data on 30% data
    
    #generate report
    print(classification_report(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    
    return model
```

```python
#call the inbuilt class 
from sklearn.metrics import classification_report , confusion_matrix
```

```python
#call the inbuilt class 
from sklearn.metrics import classification_report , confusion_matrix



#1. Classification Alogithm : Base Line Model 
#means Logistic Regression 
from sklearn.linear_model import LogisticRegression

#create an object for LogisticRegreesion class
lr = LogisticRegression()

#call the function
lr = create_model(lr)
```

```python
#2.Classifiaction Algorithm : DecisionTreeClassifier by default it takes GINI Index
#create an object for DecisionTreeClassifier class
from sklearn.tree import DecisionTreeClassifier

#create an object for DecisionTreeClassifier class
dtc = DecisionTreeClassifier(random_state=1)

#call the function
dtc = create_model(dtc)
```

```python
#To check the Information Gain :
dict1 = {"Input":x.columns,"IG":dtc.feature_importances_}
df2 = pd.DataFrame(dict1)
df2.sort_values("IG",ascending=False,ignore_index=True)
```

```python
#To Draw the tree , we call the inbuilt method tree from sklearn package
from sklearn import tree

f = x.columns
plt.figure(figsize=(10,10))
_ = tree.plot_tree(dtc,feature_names=f,filled=True)
plt.show()
```

```python
#Here we got the recall score 0.81(81%) which is good , but not better 
#So for better score we apply pruning technique on our dataset
#There are 2 types of Pruning technique
#1.Max Depth : get the best between >=1 to <=8
#2.Min Samples Leaf : get the best score between >=45 to <=100
```

```python
#1.max_depth : applying for-loop to check the best score
for i in range(1,9):
    dtc1 = DecisionTreeClassifier(random_state=1,max_depth=i)
    print("Max Depth :",i)
    #call the function
    dtc1 = create_model(dtc1)
```

```python
#Got the best score at 3 , to check
dtc1 = DecisionTreeClassifier(random_state=1,max_depth=3)
#call the function
dtc1 = create_model(dtc1)
```

```python
#To check the Information Gain :
dict2 = {"Input":x.columns,"IG":dtc1.feature_importances_}
df3 = pd.DataFrame(dict2)
df3.sort_values("IG",ascending=False,ignore_index=True)
```

```python
#2.Min Samples Leaf : 2nd pruning technique apply for loop check the best score
for i in range(45,101):
    dtc2 = DecisionTreeClassifier(random_state=1,min_samples_leaf=i)
    print("Min Samples Leaf :",i)
    #call the function
    dtc2 = create_model(dtc2)
```

```python
#Get the good score at 45 , to show
dtc2 = DecisionTreeClassifier(random_state=1,min_samples_leaf=45)
#call the function
dtc2 = create_model(dtc2)
```
